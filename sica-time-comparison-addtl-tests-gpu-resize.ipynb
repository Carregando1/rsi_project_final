Part 1: Dataset Generation

Step 1: Import relevant state functions and packages

```python
from cax_sica.genetic.state import CurrentState as st2

import numpy as np
import jax.numpy as jnp
import time
from typing import List, Tuple
```

Part 2: Testing Dim/Num Timesteps against Time Required for SICA Simulation: Trial 1: Dim Testing

```python
#Requires about 35 minutes of runtime.
data: List[Tuple] = []
iter = 11
def generateicsrt(dim, num_steps, example = 0):
    ic = np.random.choice([0,0,0,1], size=(dim, dim))
    srt = np.random.choice([0,0,0,0,0,1], size=(num_steps, dim, dim, 18))
    return ic, srt
for i in range(100, 510, 5):
    print(f"OPERATING DIM: {i}")
    for _ in range(iter):
        print(f"ITER {_}")
        stepdim = [i,100,0]
        icsrt = generateicsrt(stepdim[0], stepdim[1])
        ic = icsrt[0]
        srt = icsrt[1]
        time_init = time.time()
        state = st2(initial=ic, rules=srt)
        st2res = state.generate()
        data.append((i,time.time()-time_init))


```

Trial 2: Timestep Testing


```python
#Requires about 40 minutes of runtime.
data_time: List[Tuple] = []
iter = 11
def generateicsrt(dim, num_steps, example = 0):
    ic = np.random.choice([0,0,0,1], size=(dim, dim))
    srt = np.random.choice([0,0,0,0,0,1], size=(num_steps, dim, dim, 18))
    return ic, srt
for i in range(100, 510, 5):
    print(f"OPERATING TIMESTEP CT: {i}")
    for _ in range(iter):
        print(f"ITER {_}")
        stepdim = [200,i,0]
        icsrt = generateicsrt(stepdim[0], stepdim[1])
        ic = icsrt[0]
        srt = icsrt[1]
        time_init = time.time()
        state = st2(initial=ic, rules=srt)
        st2res = state.generate()
        data_time.append((i,time.time()-time_init))


```

Part 2: Data Analysis and Plotting

Step 1: Import relevant functions

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.metrics import r2_score
```

Step 2: Generate plot arrays

```python
#Adjust these values to be 810 - number of outliers.

dim_x = np.zeros((820))
dim_y = np.zeros((820))
step_x = np.zeros((820))
step_y = np.zeros((820))

#Remove first trials as they are outliers, as well as trial-specific outliers (dim: > 5.5s, step: > 2s)
dim_out = 0
step_out = 0
for i in range(0, len(data)):
    if (i%11):
        if (data[i][1] < 55):
            dim_x[i-1-int(np.floor(i/11)) - dim_out] = data[i][0]
            dim_y[i-1-int(np.floor(i/11)) - dim_out] = data[i][1]
        else:
            dim_out += 1
for i in range(0, len(data_time)):
    if (i%11):
        if (data_time[i][1] < 22):
            step_x[i-1-int(np.floor(i/11)) - step_out] = data_time[i][0]
            step_y[i-1-int(np.floor(i/11)) - step_out] = data_time[i][1]
        else:
            step_out += 1
```

Step 3: Perform test regressions (target: dim: quadratic; timestep: linear) and plot; evaluate using R²

```python
dim = stats.linregress(dim_x, dim_y)
step = stats.linregress(step_x, step_y)

"""Dim plots."""
#O(n) fit
plt.figure().set_figheight(3)
plt.plot(dim_x, dim_y, 'o')
plt.plot(dim_x, dim.intercept + dim.slope*dim_x, 'r')
plt.xlabel(f"Dim \n\n R-squared value: {dim.rvalue**2:.6}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Dimension Simulation: Linear Fit")
plt.show()

#O(n log n) fit
dimn15 = np.poly1d(np.polyfit((dim_x * np.log(dim_x)), dim_y, 1))
pdimn15 = np.linspace(100, 500, 1000)
plt.figure().set_figheight(3)
plt.plot(dim_x, dim_y, 'o')
plt.plot(pdimn15, dimn15(pdimn15*np.log(pdimn15)), 'r')
plt.xlabel(f"Dim \n\n R-squared value: {r2_score(dim_y, dimn15(dim_x*np.log(dim_x))):.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Dimension Simulation: Linear-Logarithmic Fit")
plt.show()

#O(n²) fit
dimn2 = np.poly1d(np.polyfit(dim_x, dim_y, 2))
pdimn2 = np.linspace(100, 500, 1000)
plt.figure().set_figheight(3)
plt.plot(dim_x, dim_y, 'o')
plt.plot(pdimn2, dimn2(pdimn2), 'r')
plt.xlabel(f"Dim \n\n R-squared value: {r2_score(dim_y, dimn2(dim_x)):.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Dimension Simulation: Quadratic Fit")
plt.show()

#O(n² log n) fit
dimn25 = np.poly1d(np.polyfit((dim_x * dim_x * np.log(dim_x)), dim_y, 1))
pdimn25 = np.linspace(100, 500, 1000)
plt.figure().set_figheight(3)
plt.plot(dim_x, dim_y, 'o')
plt.plot(pdimn25, dimn25(pdimn25*pdimn25*np.log(pdimn25)), 'r')
plt.xlabel(f"Dim \n\n R-squared value: {r2_score(dim_y, dimn25(dim_x*dim_x*np.log(dim_x))):.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Dimension Simulation: Quadratic-Logarithmic Fit")
plt.show()

#O(n^3) fit
dimn3 = np.poly1d(np.polyfit(dim_x, dim_y, 3))
pdimn3 = np.linspace(100, 500, 1000)
plt.figure().set_figheight(3)
plt.plot(dim_x, dim_y, 'o')
plt.plot(pdimn3, dimn3(pdimn3), 'r')
plt.xlabel(f"Dim \n\n R-squared value: {r2_score(dim_y, dimn3(dim_x)):.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Dimension Simulation: Cubic Fit")
plt.show()

"""Timestep plots."""
#O(n) fit
plt.figure().set_figheight(3)
plt.plot(step_x, step_y, 'o', label='original data')
plt.plot(step_x, step.intercept + step.slope*step_x, 'r')
plt.xlabel(f"Number of SICA timesteps simulated \n\n R-squared value: {step.rvalue**2:.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Timestep Simulation: Linear Fit")
plt.show()

#O(n log n) fit
timn15 = np.poly1d(np.polyfit((step_x * np.log(step_x)), step_y, 1))
ptimn15 = np.linspace(100, 500, 1000)
plt.figure().set_figheight(3)
plt.plot(step_x, step_y, 'o')
plt.plot(ptimn15, timn15(ptimn15*np.log(ptimn15)), 'r')
plt.xlabel(f"Number of SICA timesteps simulated \n\n R-squared value: {r2_score(step_y, timn15(step_x*np.log(step_x))):.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Timestep Simulation: Linear-Logarithmic Fit")
plt.show()

#O(n²) fit
timn2 = np.poly1d(np.polyfit(step_x, step_y, 2))
ptimn2 = np.linspace(100, 500, 1000)
plt.figure().set_figheight(3)
plt.plot(step_x, step_y, 'o')
plt.plot(ptimn2, timn2(ptimn2), 'r')
plt.xlabel(f"Number of SICA timesteps simulated \n\n R-squared value: {r2_score(step_y, timn2(step_x)):.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Timestep Simulation: Quadratic Fit")
plt.show()

#O(n² log n) fit
timn25 = np.poly1d(np.polyfit((step_x * step_x * np.log(step_x)), step_y, 1))
ptimn25 = np.linspace(100, 500, 1000)
plt.figure().set_figheight(3)
plt.plot(step_x, step_y, 'o')
plt.plot(ptimn25, timn25(ptimn25*ptimn25*np.log(ptimn25)), 'r')
plt.xlabel(f"Number of SICA timesteps simulated \n\n R-squared value: {r2_score(step_y, timn25(step_x*step_x*np.log(step_x))):.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Timestep Simulation: Quadratic-Logarithmic Fit")
plt.show()

#O(n^3) fit
timn3 = np.poly1d(np.polyfit(step_x, step_y, 3))
ptimn3 = np.linspace(100, 500, 1000)
plt.figure().set_figheight(3)
plt.plot(step_x, step_y, 'o')
plt.plot(ptimn3, timn3(ptimn3), 'r')
plt.xlabel(f"Number of SICA timesteps simulated \n\n R-squared value: {r2_score(step_y, timn3(step_x)):.6f}")
plt.ylabel("Runtime (seconds)")
plt.title("GPU Timestep Simulation: Cubic Fit")
plt.show()
```
